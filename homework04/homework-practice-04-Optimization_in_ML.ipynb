## 1.2 –ö—Ä–∏—Ç–µ—Ä–∏–π –æ—Å—Ç–∞–Ω–æ–≤–∫–∏

$$ ‚Äñ‚àáf(x_k)‚Äñ^2_2 ‚â§ Œµ‚Äñ‚àáf(x_0)‚Äñ^2_2 \tag{1.1},$$
–≥–¥–µ $Œµ‚àà(0,1)$ - –∑–∞–¥–∞–Ω–Ω–∞—è –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è—Ç–æ—á–Ω–æ—Å—Ç—å. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –∫—Ä–∏—Ç–µ—Ä–∏–π –æ—Å—Ç–∞–Ω–æ–≤–∫–∏ (1.1) –≥–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç, —á—Ç–æ –º–µ—Ç–æ–¥ —É–º–µ–Ω—å—à–∏—Ç –Ω–∞—á–∞–ª—å–Ω—É—é –Ω–µ–≤—è–∑–∫—É $‚Äñ‚àáf(x_0)‚Äñ_2$ –≤ $Œµ^{‚àí1}$ —Ä–∞–∑. –í —ç—Ç–æ–º –∑–∞–¥–∞–Ω–∏–∏ –í–∞–º –Ω—É–∂–Ω–æ –±—É–¥–µ—Ç –≤–æ –≤—Å–µ—Ö –º–µ—Ç–æ–¥–∞—Ö –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∫—Ä–∏—Ç–µ—Ä–∏–π –æ—Å—Ç–∞–Ω–æ–≤–∫–∏ (1.1).

## 1.3 –õ–∏–Ω–µ–π–Ω—ã–π –ø–æ–∏—Å–∫

$$œÜ‚Ä≤_k(Œ±) =„Äà‚àáf(x_k+Œ±d_k), d_k„Äâ.$$

**–ê–ª–≥–æ—Ä–∏—Ç–º 2** –ú–µ—Ç–æ–¥ –¥—Ä–æ–±–ª–µ–Ω–∏—è —à–∞–≥–∞

**–í—ã—Ö–æ–¥:** $Œ±$  

"–ê–¥–∞–ø—Ç–∏–≤–Ω—ã–π" –º–µ—Ç–æ–¥ –ø–æ–¥–±–æ—Ä–∞ —à–∞–≥–∞ –∑–∞–ø–æ–º–∏–Ω–∞–µ—Ç –≤–µ–ª–∏—á–∏–Ω—É $Œ±_k$, –Ω–∞–π–¥–µ–Ω–Ω—É—é –Ω–∞ —Ç–µ–∫—É—â–µ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ –∏

## 1.5 –ú–µ—Ç–æ–¥ –ù—å—é—Ç–æ–Ω–∞

## –ü—Ä–æ–≤–µ—Ä–∏–º –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç—å —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –ø–æ–¥—Å—á–µ—Ç–∞ –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞ –∏ –≥–µ—Å—Å–∏–∞–Ω–∞.
 –ü—Ä–æ–≤–µ—Ä–∏–º –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç—å —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ –ø–æ–¥—Å—á–µ—Ç–∞ –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞ –∏ –≥–µ—Å—Å–∏–∞–Ω–∞ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ä–∞–∫—É–ª–∞ —Å –ø–æ–º–æ—â—å—é —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π `grad_finite_diff` –∏ `hess_finite_diff` –≤ –º–æ–¥—É–ª–µ oracles. –î–ª—è —ç—Ç–æ–≥–æ —Å–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –Ω–µ–±–æ–ª—å—à—É—é –º–æ–¥–µ–ª—å–Ω—É—é –≤—ã–±–æ—Ä–∫—É (–º–∞—Ç—Ä–∏—Ü—É  ùê¥ –∏ –≤–µ–∫—Ç–æ—Ä  ùëè) –∏ —Å—Ä–∞–≤–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏—è, –≤—ã–¥–∞–≤–∞–µ–º—ã–µ –º–µ—Ç–æ–¥–∞–º–∏ `grad` –∏ `hess`, —Å —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–º–∏ —Ä–∞–∑–Ω–æ—Å—Ç–Ω—ã–º–∏ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏—è–º–∏.

````python
import numpy as np
from oracles import create_log_reg_oracle, LogRegL2Oracle, grad_finite_diff, hess_finite_diff
from scipy.sparse import diags

def accuracy(func, func_finite):
    accur = abs(func - func_finite)
    return accur

np.random.seed(42)
n_samples, n_features = 100, 3
A = np.random.randn(20, 3)
b = np.random.choice([1, -1], size=20)

regcoef = 0.1

oracle = create_log_reg_oracle(A, b, regcoef)
x = np.array([1., 2., 3.])

print(f'Gradient: {oracle.grad(x)}\n\nGradient finite: {grad_finite_diff(oracle.func, x)}\n\n\
Hessian:\n{oracle.hess(x)}\n\nHessian finite:\n{hess_finite_diff(oracle.func, x)}\n')

print(f'Gradient accuracy: {accuracy(oracle.grad(x), grad_finite_diff(oracle.func, x))}\n\n\
Hessian accuracy:\n {accuracy(oracle.hess(x), hess_finite_diff(oracle.func, x))}')
````

–î–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è –ª–∏–Ω–∏–π —É—Ä–æ–≤–Ω—è –º–æ–∂–µ—Ç–µ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —Ñ—É–Ω–∫—Ü–∏–µ–π `plot_levels`, –∞ –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è
—Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–π `plot_trajectory` –∏–∑ —Ñ–∞–π–ª–∞ `plot_trajectory_2d.py`, –ø—Ä–∏–ª–∞–≥–∞—é—â–µ–≥–æ—Å—è –∫ –∑–∞–¥–∞–Ω–∏—é.  
–¢–∞–∫–∂–µ –æ–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –æ—Ä–∞–∫—É–ª –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ `QuadraticOracle` —É–∂–µ —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω –≤
–º–æ–¥—É–ª–µ `oracles`. –û–Ω —Ä–µ–∞–ª–∏–∑—É–µ—Ç —Ñ—É–Ω–∫—Ü–∏—é $f(x) = (1/2)„ÄàAx, x„Äâ‚àí„Äàb, x„Äâ$, –≥–¥–µ $A‚ààS^n_{++}, b ‚àà R^n$.



````python
from plot_trajectory_2d import plot_levels, plot_trajectory
from optimization import gradient_descent, newton, LineSearchTool
import oracles as orl
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
````

–†–µ–∞–ª–∏–∑—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é `gradient_trajectory` –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ –∏ —Ä–∏—Å–æ–≤–∞–Ω–∏—è –ª–∏–Ω–∏–π –∏ —Ç—Ä–∞–µ–∫—Ç–æ—Ä–∏–π.

````python
def gradient_trajectory(A, b, x_0, methods):
    
    fig, ax = plt.subplots(3, 3, figsize=(15, 15))
    oracle = orl.QuadraticOracle(A, b)
    
    for i in range(3):
        for j in range(3):
            if methods.get('method')[i] == 'Constant':
                line_search_tool = {'method': methods.get('method')[i], 'c': methods.get('c1')[j]}
                
            elif methods.get('method')[i] == 'Armijo':
                line_search_tool = {'method': methods.get('method')[i], 'c1': methods.get('c1')[j]}
                
            elif methods.get('method')[i] == 'Wolfe':
                line_search_tool = {'method': methods.get('method')[i], 'c1': methods.get('c1')[j],
                                    'c2' : methods.get('c2')[j]}
            
            _, _, history = gradient_descent(oracle, x_0, line_search_options=line_search_tool, trace=True)
            k = len(history['x'])
            ax[i][j].set_title(f'{line_search_tool} iter: {k}')
            plot_levels(oracle.func, ax[i][j])
            plot_trajectory(oracle.func, ax[i][j], history['x'])
````

````python
A = np.array([[3., 0.], [0., 3.]])
b = np.zeros(2)
print('–û–±—É—Å–ª–æ–≤–ª–µ–Ω–Ω–æ—Å—Ç—å', np.linalg.cond(A))

x_0 = [5., -2.]

methods = {'method': ['Constant', 'Wolfe', 'Armijo'], 'c1': [0.1, 0.25, 0.5], 'c2': [0.1, 0.5, 1.]}

gradient_trajectory(A, b, x_0, methods)
````

### –î–ª—è —Ö–æ—Ä–æ—à–æ –æ–±—É—Å–ª–æ–≤–ª–µ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –ª—É—á—à–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –ø–æ–∫–∞–∑–∞–ª –º–µ—Ç–æ–¥ –í—É–ª—å—Ñ–∞.

````python
A = np.array([[2., 1.9],
             [1.9, 2.1]])
b = np.zeros(2)
print('–û–±—É—Å–ª–æ–≤–ª–µ–Ω–Ω–æ—Å—Ç—å', np.linalg.cond(A))

x_0 = [3., 2.]

gradient_trajectory(A, b, x_0, methods)
````

**–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è:** –ü—Ä–æ—â–µ –≤—Å–µ–≥–æ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å —Å–ª—É—á–∞–π–Ω—É—é –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω—É—é –∑–∞–¥–∞—á—É —Ä–∞–∑–º–µ—Ä–∞ $n$ —Å –∑–∞–¥–∞–Ω–Ω—ã–º —á–∏—Å–ª–æ–º –æ–±—É—Å–ª–æ–≤–ª–µ–Ω–Ω–æ—Å—Ç–∏ $Œ∫$ —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º. –í –∫–∞—á–µ—Å—Ç–≤–µ –º–∞—Ç—Ä–∏—Ü—ã $A‚ààS^n_{++}$ —É–¥–æ–±–Ω–æ –≤–∑—è—Ç—å –ø—Ä–æ—Å—Ç–æ –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É $A= Diag(a)$, —É –∫–æ—Ç–æ—Ä–æ–π –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω—ã —Å–ª—É—á–∞–π–Ω–æ

–¢–µ–ø–µ—Ä—å —É–≤–µ–ª–∏—á—å—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ $n$ –∏ –ø–æ–≤—Ç–æ—Ä–∏—Ç–µ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç —Å–Ω–æ–≤–∞. –í—ã –¥–æ–ª–∂–Ω—ã –ø–æ–ª—É—á–∏—Ç—å –Ω–æ–≤–æ–µ —Å–µ–º–µ–π—Å—Ç–≤–æ
–∫—Ä–∏–≤—ã—Ö $T(n‚Ä≤,Œ∫)$ –ø—Ä–æ—Ç–∏–≤ $Œ∫$. –ù–∞—Ä–∏—Å—É–π—Ç–µ –∏—Ö –≤—Å–µ –æ–¥–Ω–∏–º –∏ —Ç–µ–º –∂–µ —Ü–≤–µ—Ç–æ–º, –Ω–æ –æ—Ç–ª–∏—á–Ω—ã–º –æ—Ç –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ
(–Ω–∞–ø—Ä–∏–º–µ—Ä, —Å–∏–Ω–∏–º).  

–ü–æ–≤—Ç–æ—Ä–∏—Ç–µ —ç—Ç—É –ø—Ä–æ—Ü–µ–¥—É—Ä—É –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –¥–ª—è –¥—Ä—É–≥–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π $n$. –í –∏—Ç–æ–≥–µ –¥–æ–ª–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å—Å—è –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑–Ω—ã—Ö —Å–µ–º–µ–π—Å—Ç–≤ –∫—Ä–∏–≤—ã—Ö - —á–∞—Å—Ç—å –∫—Ä–∞—Å–Ω—ã—Ö (—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –æ–¥–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é $n$), —á–∞—Å—Ç—å —Å–∏–Ω–∏—Ö (—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –¥—Ä—É–≥–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é $n$), —á–∞—Å—Ç—å –∑–µ–ª–µ–Ω—ã—Ö –∏ —Ç. –¥.  

–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –∑–Ω–∞—á–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ $n$ –∏–º–µ–µ—Ç —Å–º—ã—Å–ª –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å –ø–æ –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–æ–π
—Å–µ—Ç–∫–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, $n = 10, n = 100, n = 1000$ –∏ —Ç. –¥.).  

–ö–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –∏–∑ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –∫–∞—Ä—Ç–∏–Ω–∫–∏?

````python
def matrix_generate(k, n):
    diag = np.random.uniform(1, k, n)
    diag[0] = 1.0
    diag[n-1] = k
    A = diags(diag)
    
    return A

n = [10, 100, 1000, 10000]
k = [1.0, 20.0, 50.0, 80.0, 100.0]
colors = ['blue', 'green', 'red', 'yellow']

for i in range(len(n)):
    count_iter = np.zeros(len(k))
    count_cond = np.zeros(len(k))
    
    for m in range(100):
        for j in range(len(k)):
            A = matrix_generate(k[j], n[i])
            b = np.random.uniform(0, 1, n[i]) * k[j]
            oracle = orl.QuadraticOracle(A, b)
            [x_star, msg, history] = gradient_descent(oracle=oracle, x_0=np.random.uniform(0.5, 1, n[i]),
                                                                       trace=True)
            count_iter[j] += len(history['func'])
            count_cond[j] = k[j]
    
    plt.plot(count_cond, count_iter / 100, c=colors[i], label=(f'n = {n[i]}'))
    
plt.xticks(k)
plt.xlabel("Number of conditionality")
plt.ylabel("Number of iterations")
plt.title("Dependence of the number of iterations on the number of conditionality and dimension of space")
plt.legend()
plt.show()
````

## 3.3 –≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç: –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç–æ–¥–æ–≤ –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞ –∏ –ù—å—é—Ç–æ–Ω–∞ –Ω–∞ —Ä–µ–∞–ª—å–Ω–æ–π –∑–∞–¥–∞—á–µ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏

—Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.

–í –∫–∞—á–µ—Å—Ç–≤–µ —Ä–µ–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ —Ç—Ä–∏ –Ω–∞–±–æ—Ä–∞ —Å —Å–∞–π—Ç–∞ LIBSVM [http://www.csie.ntu.edu.tw/~cjlin/libsvmtools/datasets/.](http://www.csie.ntu.edu.tw/~cjlin/libsvmtools/datasets/.)
: *w8a*, *gisette* –∏ *real-sim*. –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏ –≤–∑—è—Ç—å —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –æ–±—Ä–∞–∑–æ–º: $Œª = 1/m$.
–ü–∞—Ä–∞–º–µ—Ç—Ä—ã –æ–±–æ–∏—Ö –º–µ—Ç–æ–¥–æ–≤ –≤–∑—è—Ç—å —Ä–∞–≤–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é. –ù–∞—á–∞–ª—å–Ω—É—é —Ç–æ—á–∫—É –≤—ã–±—Ä–∞—Ç—å
$x_0 = 0$.

–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ —Å—Ö–æ–¥–∏–º–æ—Å—Ç–∏ —Å–ª–µ–¥—É—é—â–∏—Ö –¥–≤—É—Ö –≤–∏–¥–æ–≤:  
1) –ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ –æ—Ç —Ä–µ–∞–ª—å–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ —Ä–∞–±–æ—Ç—ã –º–µ—Ç–æ–¥–∞.  
2) –ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–∞ –Ω–æ—Ä–º—ã –≥—Ä–∞–¥–∏–µ–Ω—Ç–∞ $‚Äñ‚àáf(x_k)‚Äñ^2_2 /‚Äñ‚àáf(x_0)‚Äñ^2_2$ (–≤ –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–æ–π —à–∫–∞–ª–µ) –ø—Ä–æ—Ç–∏–≤ —Ä–µ–∞–ª—å–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ —Ä–∞–±–æ—Ç—ã.

–ü—Ä–∏ —ç—Ç–æ–º –æ–±–∞ –º–µ—Ç–æ–¥–∞ (–≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫ –∏ –ù—å—é—Ç–æ–Ω) –Ω—É–∂–Ω–æ —Ä–∏—Å–æ–≤–∞—Ç—å –Ω–∞ –æ–¥–Ω–æ–º –∏ —Ç–æ–º –∂–µ –≥—Ä–∞—Ñ–∏–∫–µ.
–£–∫–∞–∂–∏—Ç–µ –≤ –æ—Ç—á–µ—Ç–µ, –∫–∞–∫–æ–≤–∞ —Å—Ç–æ–∏–º–æ—Å—Ç—å –∏—Ç–µ—Ä–∞—Ü–∏–∏ –∏ —Å–∫–æ–ª—å–∫–æ –ø–∞–º—è—Ç–∏ —Ç—Ä–µ–±—É–µ—Ç—Å—è –∫–∞–∂–¥–æ–º—É –∏–∑ –º–µ—Ç–æ–¥–æ–≤ –≤
–∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $m$ (—Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏) –∏ $n$ (—Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–∞). –ü—Ä–∏ –æ—Ü–µ–Ω–∫–µ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –Ω–æ—Ç–∞—Ü–∏—é $O(¬∑)$, —Å–∫—Ä—ã–≤–∞—é—â—É—é –≤–Ω—É—Ç—Ä–∏ —Å–µ–±—è –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—ã.

–ö–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –ø–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º —ç—Ç–æ–≥–æ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞? –ö–∞–∫–æ–π –∏–∑ –º–µ—Ç–æ–¥–æ–≤ –ª—É—á—à–µ –∏ –≤
–∫–∞–∫–∏—Ö —Å–∏—Ç—É–∞—Ü–∏—è—Ö?

**–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏—è:** –õ—é–±–æ–π –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö —Å —Å–∞–π—Ç–∞ LIBSVM –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç –∏–∑ —Å–µ–±—è —Ç–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª –≤ —Ñ–æ—Ä-
–º–∞—Ç–µ svmlight. –ß—Ç–æ–±—ã —Å—á–∏—Ç–∞—Ç—å —Ç–∞–∫–æ–π —Ç–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª, –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é `load_svmlight_file` –∏–∑ –º–æ–¥—É–ª—è `sklearn.datasets`. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ —ç—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –º–∞—Ç—Ä–∏—Ü—É –≤ —Ñ–æ—Ä–º–∞—Ç–µ `scipy.sparse.csr_matrix`, –ø–æ—ç—Ç–æ–º—É –í–∞—à–∞ —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ä–∞–∫—É–ª–∞ –¥–æ–ª–∂–Ω–∞ –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞—Ç—å —Ç–∞–∫–∏–µ –º–∞—Ç—Ä–∏—Ü—ã.

````python
from sklearn.datasets import load_svmlight_file

def grad_norm(history):
    return np.array(history['grad_norm']) ** 2 / history['grad_norm'][0] ** 2

def graph(newton, gradient):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    ax[0].plot(gradient['time'], gradient['func'], label='Gradient descent')
    ax[0].plot(newton['time'], newton['func'], label='Newtons method')
    ax[0].set_ylabel("Value of function")
    ax[0].set_xlabel("Time")
    ax[0].legend()

    ax[1].plot(gradient['time'], grad_norm(gradient), label='Gradient descent')
    ax[1].plot(newton['time'], grad_norm(newton), label='Newtons method')
    ax[1].set_ylabel("$‚Äñ‚àáf(x_k)‚Äñ^2_2 /‚Äñ‚àáf(x_0)‚Äñ^2_2$")
    ax[1].set_xlabel("Time")
    ax[1].set_yscale("log")
    ax[1].legend()
````

````python
A, b = load_svmlight_file('data/w8a.txt')

m = A.shape[0]
n = A.shape[1]
regcoef = 1 / m
x_0 = np.zeros(n)

oracle = orl.create_log_reg_oracle(A, b, regcoef)
[x_star_grad, msg_grad, gradient] = gradient_descent(oracle=oracle, x_0=x_0, trace=True)
[x_star_newton, msg_newton, newton] = newton(oracle=oracle, x_0=x_0, trace=True)

graph(newton, gradient)

newton_iter = len(newton['func'])
grad_iter = len(gradient['func'])
print(f'Iterations of Gradient descent: {grad_iter}\nIterations of Newtons method: {newton_iter}')
````
